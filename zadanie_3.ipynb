Import potrzebnych pakietow:

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Zdefiniowanie funkcji:

```python
def f(x):
    return x*x + 5
```

Definicja trzech wspolrzednych x i y z przykladow 1-3:

```python
x_1 = np.linspace(-1, 1, 100)
y_1 = f(x_1)
x_2 = np.linspace(-6, 6, 600)
y_2 = f(x_2)
x_3 = np.linspace(0, 5, 250)
y_3 = f(x_3)
```

Funkcja odpowiedzialna za sporzadzenie wykresu wraz z opisem:

```python
def plot(x, y, y_axis_low, y_axis_high):
    fig, ax = plt.subplots()
    ax.plot(x, y)
    plt.title("Funkcja f(x) = x^2 + 5")
    plt.xlabel("x")
    plt.ylabel("y")
    #plt.xlim([-x, x])
    plt.ylim([y_axis_low, y_axis_high])
    plt.show()
```

Pokazanie wykresow:

```python
plot(x_1, y_1, 4.5, 6)
plot(x_2, y_2, 0, 30)
plot(x_3, y_3, 0, 30)
```
