Import potrzebnych pakietow:

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Zdefiniowanie funkcji:

```python
def f(x):
    return x*x + 5
```

Definicja trzech wspolrzednych x i y z przykladow 1-3 z pliku README.md:

```python
x_1 = np.linspace(-1, 1, 100)
y_1 = f(x_1)
x_2 = np.linspace(-6, 6, 600)
y_2 = f(x_2)
x_3 = np.linspace(0, 5, 250)
y_3 = f(x_3)
```

Funkcja odpowiedzialna za sporzadzenie wykresu wraz z opisem:

```python
def plot(x, y, y_axis_low, y_axis_high, number):
    fig, ax = plt.subplots()
    ax.plot(x, y)
    plt.title("Funkcja f(x_{}) = x_{}^2 + 5".format(number, number))
    plt.xlabel("x_{}".format(number))
    plt.ylabel("f(x_{}) = y_{}".format(number, number))
    #plt.xlim([-x, x])
    plt.ylim([y_axis_low, y_axis_high])
    plt.show()
```

Pokazanie wykresow:

```python
plot(x_1, y_1, 4.5, 6, 1)
plot(x_2, y_2, 0, 30, 2)
plot(x_3, y_3, 0, 30, 3)
```
